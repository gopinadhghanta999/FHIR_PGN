```python
spark.sql("CREATE CATALOG IF NOT EXISTS fhir_data")
spark.sql("CREATE SCHEMA IF NOT EXISTS fhir_data.prd_raw")
spark.sql("CREATE SCHEMA IF NOT EXISTS fhir_data.prd_bronze")
spark.sql("CREATE SCHEMA IF NOT EXISTS fhir_data.prd_silver")
spark.sql("CREATE SCHEMA IF NOT EXISTS fhir_data.prd_gold")
spark.sql("CREATE VOLUME if not exists fhir_data.prd_raw.firstv")
```

```python
CONFIG = {
    "resources": ["Patient", "Encounter", "Observation", "Condition"],
    "base_path": "/Volumes/fhir_data/prd_raw/firstv/",
    "raw_path": "/Volumes/fhir_data/prd_raw/firstv/raw",
    "num_days" :3
}   
```

```python
def create_fld_str(config):
    layers = ['raw']

    for layer in layers:
        for resource in config['resources']:
            path = f"{config['base_path']}/{layer}/{resource.lower()}"

            try:
                dbutils.fs.ls(path)
                print(f"Already exists: {path}")
            except:
                dbutils.fs.mkdirs(path)
                print(f"Created: {path}")

create_fld_str(CONFIG)
```
